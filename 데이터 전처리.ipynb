```python
## case 2 : 자동화를 위한 merge 함수 제작

def merge(df, region_old, region_new): 
    #데이터프레임, 바꿀 지역명, 새 지역명
    #도시를 하나씩 넣거나, 리스트로 한꺼번에 넣어 구분함
    if isinstance(region_old, str) and isinstance(region_new, str): 
        ro = region_old
        rn = region_new
        col = list(df.filter(regex=ro)) 
        #바꿀 지역명을 포함한 모든 col 을 list로 반환
        df[rn] = df[col].sum(axis=1)
        # 새로운 지역명으로 col 변수를 추가
        # 해당 col의 값은 원래 지역명들의 값을 합한 값
        df.drop(columns = col, inplace=True)
        # 원래 존재하던 여러 개의 지역명 제거
        return df

    elif isinstance(region_old, list) and isinstance(region_new, list):
        # 리스트에서 하나씩 꺼내서 작업
        for i, ro in enumerate(region_old):
            col = list(df.filter(regex=(ro)))
            df[region_new[i]] =  df[col].sum(axis=1)
            df.drop(columns =  col, inplace=True)
        return df

    else :
        print("제대로 입력해주세요.")


list_old = ['경기', '전남', '강원', '충남', '전북', '제주', '경북', '충북', '경남']
# 바꾸고자 하는 지역명의 앞부분 값 list
# 경기 수원, 경기 과천 등을 경기도로 바꾸고 싶음
# 따라서 '경기' 라는 말이 들어가는 col을 선택
# 다른 지역명들도 동일

list_new = ['경기도','전라남도','강원도','충청남도','전라북도','제주도','경상북도','충청북도','경상남도']
# 바꾸어질 지역명의 이름 list

merge(crime_df2, list_old, list_new)
```
