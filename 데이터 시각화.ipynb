```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
plt.rc('font', family='NanumBarunGothic')
# 이거 먼저 설치한 후 주석처리 하고 나서 런타임 하세요.
# !sudo apt-get install -y fonts-nanum
# !sudo fc-cache -fv
# !rm ~/.cache/matplotlib -rf
sns.set(font="NanumBarunGothic", 
rc={"axes.unicode_minus":False}, style='white')
```

```python
df = pd.read_csv('data_0729.csv',  encoding='cp949', index_col=0)
df
```

```python
#제근+기나님 시각화에서 v3으로 좀 더 추가해 봤어요. trendline을 넣어주면 좀더 설명하기 좋을것 같아요.
#그리고 주요범죄율에 이미 강력범죄율과 폭력, 절도 범죄율이 정의상 추가되어서, 주요범죄율 별 지역총소득, 인구밀도, 경찰관비율,대졸자비율의 상관계수를 살펴보면 좋을것 같아요.
fig =  px.scatter(df, x=df['지역총소득'], y=df['주요범죄율'],  size= df['주요범죄율'], color=df['행정구역(시도)'], trendline='ols', trendline_scope="overall", \
                  trendline_color_override='blue')

fig.show()
#df_corr=df.corr()
```

```python
#사실 인구수 대비 주요범죄를 살펴보면 당연히 사람이 많이 살면, 주요범죄가 많이 일어나기 때문에, 인구밀도 별 주요범죄율을 살펴보는게 좋을 것 같아요. 
fig =  px.scatter(df, x=df['인구밀도'], y=df['주요범죄율'],  size= df['주요범죄율'], color=df['행정구역(시도)'], trendline='ols', trendline_scope="overall", \
                  trendline_color_override='blue')
fig.show()
```

```python
fig =  px.scatter(df, x=df['경찰관비율'], y=df['주요범죄율'],  size= df['주요범죄율'], color=df['행정구역(시도)'], trendline='ols', trendline_scope="overall", \
                  trendline_color_override='blue')
fig.show()
```

```python
fig =  px.scatter(df, x=df['대졸자비율'], y=df['주요범죄율'],  size= df['주요범죄율'], color=df['행정구역(시도)'], trendline='ols', trendline_scope="overall", \
                  trendline_color_override='blue')
fig.show()
```

```python
#마지막에 인구밀도가 적은 제주도는 유독 범죄율이 높다고 설명할 때, 이 시각화 표 하나만 설명해도 충분할 것 같아요.!
import plotly.express as px
fig = px.treemap(df, path=[df['행정구역(시도)']], values=df['인구밀도'],color_continuous_scale='YlorRd',
                 color=df['주요범죄율'])
fig.show()
```
